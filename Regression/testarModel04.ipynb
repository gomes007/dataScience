```python
import pandas as pd
import joblib

# Carrega o modelo e a lista de colunas do treino
modelo = joblib.load("modelo_score.pkl")
colunas_treino = joblib.load("colunas_treino.pkl")

# Lê novos dados
dados = pd.read_excel(r"C:\Users\Paulo\Desktop\MBA DATASCIENCE\Regression\admissao2026.xlsx")

# Pré-processa (dummies + ordena colunas)
dados_proc = pd.get_dummies(dados)
dados_proc = dados_proc.reindex(columns=colunas_treino, fill_value=0)

# Faz previsão
scores_previstos = modelo.predict(dados_proc)

# Mostra resultado junto com dados originais
dados["Score Previsto"] = scores_previstos
display(dados)

# (Opcional) Salvar em Excel
dados.to_excel("resultado_admissao2027.xlsx", index=False)
```
